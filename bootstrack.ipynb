```python
file_path = "/opt/conda/envs/boostTrack/lib/python3.8/site-packages/onnx/mapping.py"

# Đọc nội dung file và chỉnh sửa
with open(file_path, "r") as file:
    lines = file.readlines()

# Tìm và thay thế dòng có `np.object`
with open(file_path, "w") as file:
    for line in lines:
        if "np.dtype(np.object)" in line:
            # Thay thế dòng lỗi
            file.write(line.replace("np.dtype(np.object)", "np.dtype(object)"))
        else:
            file.write(line)
print(f"File {file_path} đã được chỉnh sửa thành công.")
```
